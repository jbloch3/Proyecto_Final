# Extraemos todos los archivos de los zip

```python
from zipfile import ZipFile
import os
import pandas as pd
import json
import folium
import folium.plugins
```

```python
archivos = os.listdir()
```

```python
archivos_zip = []
for i in archivos:
    if i.endswith(".zip"):
        archivos_zip.append(i)
```

```python
archivos_zip
    
```

```python
cwd = os.getcwd()
```

```python
for i in archivos_zip:
    zf = ZipFile(i)
    zf.extractall()
```

```python
df.url.tail
```

```python
for i in range(len(df.url)):
    if '//' in df.url[i]:
        df.url[i] = df.url[i].split('//')[1]
        
```
